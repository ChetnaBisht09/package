```python
import pandas as pd
import ast
df = pd.read_csv("/content/drive/MyDrive/International_T20_Data.csv")
df['info.teams']=df['info.teams'].apply(ast.literal_eval)
matches_played=pd.concat([df['info.teams'].str[0],df['info.teams'].str[1]]).value_counts()
matches_won= df['info.outcome.winner'].value_counts()
win_stats=pd.DataFrame({'matches_played':matches_played,'matches_won':matches_won}).fillna(0)
win_stats['win_percentage']=(win_stats['matches_won'])/win_stats['matches_played']*100
top5=win_stats.sort_values(by='win_percentage',ascending=False).head(5)
print(top5)
print(matches_played.head())
print(matches_won.head())
```

```python
import pandas as pd
import ast

# 1. Load the data
df=pd.read_csv("/content/drive/MyDrive/International_T20_Data.csv")

# 2. Select a match (let's pick the first match in the file, which is index 0)
match_index = 0
match_row = df.iloc[match_index]

# 3. Convert the 'innings' text into a list of Python data
# (ast.literal_eval just tells Python to treat the text as a list)
innings_list = ast.literal_eval(match_row['innings'])

# 4. Loop through each inning (usually there are 2 innings)
for inning in innings_list:
    # Get the name of the inning (e.g., '1st innings') and its data
    inning_label = list(inning.keys())[0]
    inning_data = inning[inning_label]

    team_name = inning_data['team']
    deliveries = inning_data['deliveries']

    # This dictionary will store our stats: { 'Player Name': [runs, balls] }
    batsman_stats = {}

    # 5. Look at every ball bowled in the inning
    for ball_dict in deliveries:
        for ball_number, ball_info in ball_dict.items():
            batsman = ball_info['batsman']
            runs = ball_info['runs']['batsman']

            # If the batsman is not in our dictionary yet, add them
            if batsman not in batsman_stats:
                batsman_stats[batsman] = {'Runs': 0, 'Balls': 0}

            # Update their score and ball count
            batsman_stats[batsman]['Runs'] += runs
            batsman_stats[batsman]['Balls'] += 1

    # 6. Print the Scorecard for this team
    print(f"\n--- Scorecard: {team_name} ({inning_label}) ---")
    print(f"{'Batsman':<20} {'Runs':<10} {'Balls':<10} {'SR':<10}")
    print("-" * 50)

    for player, data in batsman_stats.items():
        runs = data['Runs']
        balls = data['Balls']
        # Calculate Strike Rate (Runs per 100 balls)
        sr = (runs / balls) * 100
        print(f"{player:<20} {runs:<10} {balls:<10} {sr:.2f}")
```
